#  2. The population of each state.

Sources: http://cyberjournalist.org.in/census/cenindia.html
        https://en.wikipedia.org/wiki/List_of_states_in_India_by_past_population

* We have population information for each states for year 2001 and 2011 as per census data provided by our government.
* Comparing the years 2001 and 2011 population variation, I have filled up the population for all years, by finding the rate of increase/decrease in population for each state.
* Population rate of all states have been increased from 2001 to 2011 except NAGALAND.

```python
df_pop = pd.read_excel("Population_2001_to_2012_statewise.xlsx")
df_pop
```

```python
df_pop.columns
```

```python
df_pop.dtypes
```

```python
#Changing datatypes of Population, since it cannot be in decimal.

df_pop['2002_Population']= df_pop['2002_Population'].astype(int)
df_pop['2003_Population']= df_pop['2003_Population'].astype(int)
df_pop['2004_Population']= df_pop['2004_Population'].astype(int)
df_pop['2005_Population']= df_pop['2005_Population'].astype(int)
df_pop['2006_Population']= df_pop['2006_Population'].astype(int)
df_pop['2007_Population']= df_pop['2007_Population'].astype(int)
df_pop['2008_Population']= df_pop['2008_Population'].astype(int)
df_pop['2009_Population']= df_pop['2009_Population'].astype(int)
df_pop['2010_Population']= df_pop['2010_Population'].astype(int)
df_pop['2012_Population']= df_pop['2012_Population'].astype(int)
```
